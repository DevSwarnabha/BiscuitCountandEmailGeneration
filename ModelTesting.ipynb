```python
import cv2
image_path = r"C:\Users\Bhai\Desktop\Computer_Vision_Intern\Biscuits\total_training\frame-47.jpg"
results = model.predict(source=image_path)

# Get the annotated result and display
annotated_frame = results[0].plot()
cv2.imshow("Prediction", annotated_frame)
cv2.waitKey(0)
cv2.destroyAllWindows()

class_counts = {"Gold" : 0, "Top" : 0}
for box in results[0].boxes:
    if box.cls == 0:
        class_counts["Gold"] += 1
    else:
        class_counts["Top"] += 1

print("Number of predicted objects by class:", class_counts)
```
